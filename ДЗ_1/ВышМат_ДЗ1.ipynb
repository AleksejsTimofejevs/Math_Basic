Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
k = 0.5
plt.plot(x, np.cos(x), label = 'y(k,x)=cos(k∙x), k = 1')
plt.plot(x, np.cos(k*x), label = 'y(k,x)=cos(k∙x), k = 0.5')

plt.xlabel("Переменная x")
plt.ylabel("Функция f(x)")

plt.legend(loc="upper right")
```
